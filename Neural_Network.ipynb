**1. What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**

Deep Learning (DL) is a specialized subfield of machine learning (ML) that uses artificial neural networks with multiple layers (hence "deep") to learn and represent data. These layers are stacked in a hierarchy, allowing the model to learn complex patterns and representations automatically from raw data.

**Evolution**

1950s (The Perceptron): The concept began with Frank Rosenblatt's Perceptron, a single-layer neural network. It could solve simple linear problems but famously failed at non-linear ones (like the XOR problem), leading to the first "AI winter."

1980s (Backpropagation): The field was revived with the (re)discovery of the backpropagation algorithm, which allowed multi-layer networks to be trained. However, these models were difficult to train due to computational limits and the "vanishing gradient" problem.

2010s (The "Boom"): The modern deep learning revolution was triggered by three main factors:

- Big Data: The availability of massive datasets (like ImageNet) to train on.

- Hardware (GPUs): The use of Graphics Processing Units (GPUs) for parallel computing, making it feasible to train very deep networks in a reasonable time.

- Algorithmic Advances: New techniques like the ReLU activation function (which helped solve the vanishing gradient problem), dropout, and better optimizers (like Adam) made training more stable and effective.

**Difference from Traditional Machine Learning**

The primary difference lies in feature engineering.

**Traditional ML (e.g., SVM, Random Forest, Logistic Regression):** Requires a human expert to manually identify and extract relevant features from the raw data. The model's performance is heavily dependent on the quality of these hand-crafted features.

**Deep Learning:** Performs automatic feature learning. The multiple layers in the network learn a hierarchy of features. For example, in image recognition, the first layer might learn simple edges, the next might learn shapes (combinations of edges), and a deeper layer might learn objects (combinations of shapes).

**Traditional Machine Learning**

- Feature Engineering:
Manual, requires domain expertise.

- Data Requirement:
Performs well on small to medium datasets.

- Architecture:
Simpler, often "shallow" models.

- Hardware:
Can run on standard CPUs.

- Interpretability:
Often easier to interpret (e.g., "Decision Tree").


**Deep Learning**

- Automatic, learned from data.
- Requires large datasets to be effective.
- Complex, multi-layer ("deep") architectures.
- Computationally intensive, requires GPUs.
- Tends to be a "black box," harder to interpret.

**2. Explain the basic architecture and functioning of a Perceptron. What are its limitations?**

**Architecture**

The Perceptron is the simplest form of an artificial neural network, consisting of a single layer. Its architecture has four main components:
- Inputs ($x_1, x_2, ..., x_n$): The set of input features.
- Weights ($w_1, w_2, ..., w_n$): A value associated with each input, indicating its importance.
- Bias ($b$): An extra parameter, similar to the y-intercept in a linear equation, that allows the decision boundary to be shifted.
- Activation Function (Step Function): A function that takes the combined input and produces a final binary output (e.g., 0 or 1).

**Functioning**

The Perceptron works in two steps:
- Weighted Sum: It calculates the weighted sum of all inputs and adds the bias.$$z = (w_1 \cdot x_1 + w_2 \cdot x_2 + ... + w_n \cdot x_n) + b$$
- Activation: The result ($z$) is passed through an activation function (typically a Heaviside step function). This function makes the final decision.If $z \ge 0$, output 1 ("fire").If $z < 0$, output 0 ("do not fire").

The "learning" process involves adjusting the weights and bias based on the error of its predictions, using the Perceptron learning rule.

**Limitations**

The primary limitation of a single Perceptron is that it can only solve linearly separable problems. This means it can only classify data that can be separated by a single straight line (or a flat plane in higher dimensions).

- The XOR Problem: The most famous example of its limitation is its inability to solve the logical XOR (exclusive OR) problem. It's impossible to draw a single straight line that separates the True (1) outputs from the False (0) outputs in an XOR gate. This limitation is what led to the development of multi-layer perceptrons (MLPs), which can solve non-linear problems.

**3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.**

**Purpose of Activation Functions**

The primary purpose of an activation function is to introduce non-linearity into the neural network.

Without a non-linear activation function, a multi-layer neural network would simply be a stack of linear equations. This entire stack could be collapsed into a single, equivalent linear equation, making the "deep" network no more powerful than a simple single-layer network. Non-linearity allows the network to learn complex, non-linear patterns and boundaries in the data.

**Functions**

**Sigmoid Formula**

$\sigma(x) = \frac{1}{1 + e^{-x}}$


**Pros**

Good for output layers in binary classification (output is a probability).



**Cons**

* Vanishing Gradient Problem: Gradients are near-zero for large positive or negative inputs, which "saturates" the neuron and stops learning.


* Output is not zero-centered.


**Tanh (Hyperbolic Tangent)**

$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

**Pros**

* Zero-centered output, which can help convergence.

**Cons**

* Also suffers from the vanishing gradient problem, though less severely than Sigmoid.


**ReLU (Rectified Linear Unit)**

$f(x) = \max(0, x)$Range(0, 1)(-1, 1)[0, $\infty$)

**Pros**

* Solves vanishing gradient problem (for $x>0$).

* Computationally very fast.

**Cons**

* Causes sparsity (many neurons output 0)

* Dying ReLU Problem: Neurons can get "stuck" at 0 if the learning rate is too high, and never activate again.


* Output is not zero-centered.


**4. What is the difference between Loss function and Cost function in neural networks? Provide examples.**

While the terms "loss function" and "cost function" are often used interchangeably, there is a subtle but important technical difference:

- Loss Function: A function that measures the error (or "loss") for a single training example.
- Cost Function: A function that measures the average loss over the entire training dataset (or a batch of examples).

The goal of the optimizer is to find the model parameters (weights and biases) that minimize the cost function.

**Examples**

Scenario: A linear regression model predicting house prices.
- Prediction: $y_{\text{pred}} = wx + b$
- True Value: $y_{\text{true}}$

- Loss Function (for one sample):We could use the Squared Error (SE).$$L = (y_{\text{true}} - y_{\text{pred}})^2$$

- Cost Function (for $m$ samples):We would use the Mean Squared Error (MSE), which is the average of the individual losses.$$J = \frac{1}{m} \sum_{i=1}^{m} (y_{\text{true}}^{(i)} - y_{\text{pred}}^{(i)})^2$$

**Analogy:** The loss is the penalty for getting one question wrong on a test. The cost is the average score (the grade) for the entire test. We train the model to improve the average score, not just one specific question.

**5. What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.**

**Role of Optimizers**

An optimizer is an algorithm that adjusts the weights and biases of the neural network to minimize the cost function.

It does this by calculating the gradient (derivative) of the cost function with respect to each parameter and then taking a "step" in the opposite (negative) direction of the gradient. This process, known as backpropagation, iteratively "guides" the network toward a set of parameters that produces the lowest possible error.

**Comparison**


**Gradient Descent (Batch)**

How it works

Calculates the gradient using the entire dataset before taking one step.

**Pros**

* Simple.
* Guaranteed to converge to a global minimum (for convex problems).

**Cons**

* Extremely slow on large datasets.
* Can get stuck in local minima.


**RMSprop (Root Mean Square Prop)**

Maintains a moving average of squared gradients to adapt the learning rate for each parameter.

**Pros**

* Adaptive learning rate (faster learning in flat regions, slower in steep ones).
* Good for non-stationary (noisy) problems.*

**Cons**
* Requires tuning of its hyperparameters.

**Adam (Adaptive Moment Estimation)**

Combines the ideas of RMSprop (adaptive learning rate) and Momentum (uses a moving average of the gradient itself).

**Pros**

* Fast convergence.
* Combines the benefits of Momentum and RMSprop.
* Generally the best "out of the box" default optimizer.

**Cons**
* Can be more complex than other methods.

**6. Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.**

```python
import numpy as np

class Perceptron:
    """
    A simple single-layer perceptron.
    """
    def __init__(self, learning_rate=0.1, n_epochs=10):
        self.lr = learning_rate
        self.epochs = n_epochs
        self.weights = None
        self.bias = None

    def _step_function(self, x):
        """The Heaviside step function."""
        return np.where(x >= 0, 1, 0)

    def fit(self, X, y):
        """Train the perceptron."""
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            for idx, x_i in enumerate(X):
                # the weighted sum (linear_output)
                linear_output = np.dot(x_i, self.weights) + self.bias

                y_predicted = self._step_function(linear_output)

                # the error and update rule
                update = self.lr * (y[idx] - y_predicted)

                # update weights and bias
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        """Make predictions for new data."""
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self._step_function(linear_output)
        return y_predicted

if __name__ == "__main__":
    # Inputs for the AND gate
    X = np.array([
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ])

    # Expected outputs for the AND gate
    y = np.array([0, 0, 0, 1])

    perceptron = Perceptron(learning_rate=0.1, n_epochs=10)
    perceptron.fit(X, y)

    print("Perceptron training complete.")
    print(f"Weights: {perceptron.weights}")
    print(f"Bias: {perceptron.bias}")

    print("\n--- Testing AND Gate ---")
    print(f"0 AND 0 = {perceptron.predict([0, 0])}")
    print(f"0 AND 1 = {perceptron.predict([0, 1])}")
    print(f"1 AND 0 = {perceptron.predict([1, 0])}")
    print(f"1 AND 1 = {perceptron.predict([1, 1])}")
```

**7. Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """Sigmoid activation function"""
    return 1 / (1 + np.exp(-x))

def relu(x):
    """ReLU (Rectified Linear Unit) activation function"""
    return np.maximum(0, x)

def tanh(x):
    """Tanh (Hyperbolic Tangent) activation function"""
    return np.tanh(x)

# Generate an array of x values
x = np.linspace(-10, 10, 100)

# Calculate the y values for each function
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Create the plots
plt.figure(figsize=(15, 5))

# Plot Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()

# Plot ReLU
plt.subplot(1, 3, 2)
plt.plot(x, y_relu, label='ReLU', color='red')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.grid(True)
plt.legend()

# Plot Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, label='Tanh', color='green')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

**8. Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize the images (scale pixel values from 0-255 to 0-1)
x_train, x_test = x_train / 255.0, x_test / 255.0

# Build the neural network model
model = Sequential([
    # Flatten the 28x28 images into a 1D vector (784 features)
    Flatten(input_shape=(28, 28)),

    # Add a fully-connected (Dense) hidden layer with 128 neurons
    Dense(128, activation='relu'),

    # Add the output layer with 10 neurons (one for each digit 0-9)
    # Use 'softmax' to get probability distributions for each class
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("--- Starting Model Training ---")
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_split=0.2
)
print("--- Model Training Complete ---")

final_training_accuracy = history.history['accuracy'][-1]
print(f"\nFinal Training Accuracy (from last epoch): {final_training_accuracy:.4f}")

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test Accuracy: {test_acc:.4f}")
```

**9. Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
import matplotlib.pyplot as plt

# Load the Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

x_train, x_test = x_train / 255.0, x_test / 255.0

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_train,
    y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)

history_dict = history.history
loss = history_dict['loss']
val_loss = history_dict['val_loss']
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
epochs = range(1, len(acc) + 1)

# Plot Loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, 'bo-', label='Training loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, acc, 'bo-', label='Training accuracy')
plt.plot(epochs, val_acc, 'ro-', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Output & Interpretation**

The code will first show the training progress per epoch, then display two graphs: "Training and Validation Loss" and "Training and Validation Accuracy."

**How to Interpret the Curves:**

- Training Loss (blue line): This should always decrease. It shows how well the model is learning the training data.

- Validation Loss (red line): This shows how well the model generalizes to new, unseen data.

- Training vs. Validation Accuracy (blue/red lines): Show the model's accuracy on seen vs. unseen data.

There are three main scenarios:

- Good Fit: Both training and validation loss decrease and then plateau at a low value. The training and validation accuracy increase and plateau at a high value, with the two lines remaining close together.

- Overfitting:

  - What we see: Training loss continues to decrease, but validation loss starts to increase after a certain point. The training accuracy will be much higher than the validation accuracy.

  - Interpretation: The model has started to "memorize" the training data, including its noise, and is no longer learning general patterns. It performs poorly on new data.

- Underfitting:

   - What we see: Both training and validation loss are high and do not decrease significantly. Both accuracy scores are low.

   - Interpretation: The model is too simple (or hasn't trained long enough) to capture the patterns in the data.

**10. You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.**

**Explain your real-time data science workflow:**

- **How would you design a deep learning model (perceptron or multilayer NN)?**
- **Which activation function and loss function would you use, and why?**
- **How would you train and evaluate the model, considering class imbalance?**
- **Which optimizer would be suitable, and how would you prevent overfitting?**


**Data Science Workflow**

**1. Data Preprocessing:**
- Feature Engineering: This is critical for structured data. Create new features like transaction_amount_per_customer_avg, transactions_per_hour, etc.

- Scaling: Scale all numerical features (e.g., transaction_amount) using StandardScaler or MinMaxScaler so that no single feature dominates.

- Encoding: One-hot encode categorical features like merchant_ID and customer_location.

**2. Handling Class Imbalance:**

- This is the most important step. We cannot train on the raw, imbalanced data.

- Method: Use an oversampling technique, specifically SMOTE (Synthetic Minority Over-sampling Technique). SMOTE will create new synthetic examples of the minority "fraud" class, balancing the dataset for training.

- Alternative: Undersample the majority "legitimate" class, but this risks throwing away useful data.

**3. Train/Test Split:**

- Split the original (pre-SMOTE) dataset into training and testing sets.

- Crucially: Apply SMOTE only to the training set. The test set must remain imbalanced to reflect the real-world data and provide a true evaluation.

**Model Design (Multilayer NN)**

A Multilayer Neural Network (MLP) is the correct choice. A single perceptron cannot handle the complex, non-linear patterns of fraud.

- Input Layer: A Dense layer with an input_shape matching the number of preprocessed features.
- Hidden Layers: Start with 2-3 hidden layers, for example:
   - Dense(64, activation='relu')
   - Dense(32, activation='relu')

- Output Layer: A single neuron with a sigmoid activation:
  - Dense(1, activation='sigmoid')
  
**Activation and Loss Function**

- Activation Function (Hidden Layers): ReLU. It is computationally efficient and avoids the vanishing gradient problem, making it the standard choice.

- Activation Function (Output Layer): Sigmoid. This is essential for binary classification because it outputs a value between 0 and 1, which can be interpreted as the probability that the transaction is fraudulent.

- Loss Function: Binary Cross-Entropy (binary_crossentropy). This is the standard, mathematically correct loss function to use with a sigmoid output for a binary (0/1) classification task.

**Training and Evaluation (Imbalance)**

- Training: Train the model on the SMOTE-balanced training data.

- Evaluation: Do NOT use accuracy as an evaluation metric. A model that predicts "legitimate" every time would have 99.9% accuracy but be useless.

Use these metrics instead:

- Confusion Matrix: To see the raw counts of True/False Positives and Negatives.

- Precision: $\frac{\text{True Positives}}{\text{True Positives + False Positives}}$. (Of all transactions we flagged as fraud, how many were fraud? This is important for not angering good customers).

- Recall: $\frac{\text{True Positives}}{\text{True Positives + False Negatives}}$. (Of all actual fraud transactions, how many did we catch? This is usually the most important metric for the bank).

- F1-Score: The harmonic mean of Precision and Recall.

- AUC-ROC (Area Under the Receiver Operating Characteristic Curve): Measures the model's overall ability to distinguish between the two classes.

**Optimizer and Overfitting Prevention**

- Optimizer: Adam (adam). It is an adaptive optimizer that combines the best parts of Momentum and RMSprop. It's fast, efficient, and the best all-around default choice for most problems.

- Preventing Overfitting:

  1. Dropout: This is the most effective technique. Add Dropout(0.3) layers after each hidden Dense layer. This randomly "deactivates" 30% of neurons during training, forcing the network to learn more robust, redundant features.
  
  2. Early Stopping: Monitor the validation recall or validation loss. If it stops improving for, say, 10 epochs, automatically stop the training to get the model at its peak performance before it starts to overfit.
  
  3. L2 Regularization: Add a small penalty to the weights in the dense layers (kernel_regularizer=l2(0.001)) to keep the weights small and the model simpler.
  
  This workflow creates a robust model that is trained on balanced data and evaluated using metrics that matter for a real-world imbalanced problem.
