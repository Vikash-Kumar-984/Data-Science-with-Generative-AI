**7. What is a heatmap, and when should it be used?**

- A heatmap is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
data = sns.load_dataset("iris")

# Select only numeric columns
numeric_data = data.select_dtypes(include='number')

# Compute correlation matrix
corr = numeric_data.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

**17. Explain the difference between apply() and map() in Pandas.**

- map() is used for element-wise operations on Series.

- apply() works on both Series and DataFrames for row/column-wise operations.

```python
import pandas as pd

s = pd.Series([1, 2, 3])
print(s.map(lambda x: x * 2))  # map()

df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print(df.apply(sum))  # apply()
```

**19. How does Pandas simplify time series analysis?**

- It provides features like:

- datetime indexing

- Resampling

- Rolling windows

```python
import pandas as pd

dates = pd.date_range('20230101', periods=6)
df = pd.DataFrame({'value': [1, 2, 3, 4, 5, 6]}, index=dates)
print(df.resample('2D').sum())
```

**20. What is the role of a pivot table in Pandas?**

- A pivot table summarizes data and reshapes it using aggregation.

```python
import pandas as pd

df = pd.DataFrame({
    'Dept': ['IT', 'HR', 'IT'],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Salary': [50000, 60000, 55000]
})
pivot = df.pivot_table(index='Dept', values='Salary', aggfunc='mean')
print(pivot)
```

**1. How do you create a 2D NumPy array and calculate the sum of each row?**

- We can create a 2D array using np.array() and use np.sum(axis=1) to sum each row.
