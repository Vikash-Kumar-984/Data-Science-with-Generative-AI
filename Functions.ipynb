**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

```python
from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x*2, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))
print(reduce(lambda x, y: x + y, nums))
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers():
    numbers = list(map(int, input("Enter numbers separated by space: ").split()))
    even_sum = sum(x for x in numbers if x % 2 == 0)
    print("Sum of even numbers:", even_sum)

sum_even_numbers()
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.




```python
def print_primes():
    print("Prime numbers from 1 to 200:")
    for n in range(1, 201):
        if n < 2:
            continue
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                break
        else:
            print(n, end=' ')

print_primes()
```

5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a

n = int(input("Enter number of Fibonacci terms: "))
for num in Fibonacci(n):
    print(num, end=' ')
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

n = int(input("Enter exponent limit: "))
for power in powers_of_two(n):
    print(power, end=' ')
```

7.  Implement a generator function that reads a file line by line and yields each line as a string.



```python
def read_file_lines(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line.strip()

file_path = input("Enter file path: ")
for line in read_file_lines(file_path):
    print(line)
```

8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple..

```python
def sort_by_second_element():
    n = int(input("Enter the number of tuples: "))
    tuples_list = []

    for _ in range(n):
        x, y = map(int, input("Enter a tuple (x y): ").split())
        tuples_list.append((x, y))

    # Sort using lambda by second element
    sorted_list = sorted(tuples_list, key=lambda tup: tup[1])

    print("Sorted list:", sorted_list)

sort_by_second_element()
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


```python
def convert_c_to_f():
    celsius = list(map(float, input("Enter Celsius temperatures: ").split()))
    fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
    print("Temperatures in Fahrenheit:", fahrenheit)

convert_c_to_f()
```

10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels():
    s = input("Enter a string: ")
    result = ''.join(filter(lambda ch: ch.lower() not in 'aeiou', s))
    print("String without vowels:", result)

remove_vowels()
```
